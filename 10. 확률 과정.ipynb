### Square-root Diffusion(Cox-Ingersoll-Ross)

```python
x0 = 0.05
kappa = 3.0
theta = 0.02
sigma = 0.1
```

```python
I = 10000
M = 50
dt = T / M
def srd_euler():
    xh = np.zeros((M + 1, I))
    x1 = np.zeros_like(xh) # xh와 같은 모양의 0행렬 생성
    for t in range(1 , M + 1):
        xh[t] = (xh[t - 1]
                + kappa * (theta - np.maximum(xh[t-1], 0)) * dt
                + sigma * np.sqrt(np.maximum(xh[t-1], 0)) 
                * np.sqrt(dt) * npr.randn(I))
        x1[t] = np.maximum(xh[t], 0)
    return x1
x1 = srd_euler()
```

```python
plt.hist(x1[-1], bins = 50)
plt.xlabel('value')
plt.ylabel('frequency')
plt.grid(True)
plt.title('Simulated square-root diffusion at maturity\n(Euler scheme)')
```

```python
plt.plot(x1[:, :10], lw = 1.2)
plt.xlabel('time')
plt.ylabel('index level')
plt.grid(True)
plt.axis('tight')
plt.title('Simulated square-root diffusion paths\n(Euler scheme)')
```

```python
def srd_exact():
    x2 = np.zeros((M + 1, I))
    x2[0] = x0
    for t in range(1, M + 1):
        df = 4 * theta * kappa / (sigma**2)
        c = (sigma ** 2 * (1 - np.exp(-kappa * dt))) / (4 * kappa)
        nc = np.exp(-kappa * dt) / c * x2[t-1]
        x2[t] = c * npr.noncentral_chisquare(df, nc, size = I)
    return x2
x2 = srd_exact()
```

```python
plt.hist(x2[-1], bins=50)
plt.xlabel('value')
plt.ylabel('frequency')
plt.grid(True)
plt.title('Simulated square-root diffusion at maturity\n(exact scheme)')
```

```python
plt.plot(x2[:, :10], lw=1.5)
plt.xlabel('time')
plt.ylabel('index level')
plt.grid(True)
plt.title('Simulated square-root diffusion at maturity\n(exact scheme)')
```

### Stochastic Volatility

```python
S0 = 100
r = 0.05
v0 = 0.1
kappa = 3
theta = 0.25
sigma = 0.1
rho = 0.6
T = 1
```

```python
corr_mat = np.array([[1, rho], [rho, 1]])
cho_mat = np.linalg.cholesky(corr_mat)
```

```python
cho_mat
```

```python
M = 50
I = 10000
ran_num = npr.standard_normal((2, M + 1, I))
```

```python
dt = T / M
v = np.zeros_like(ran_num[0])
vh = np.zeros_like(v)
v[0] = v0
vh[0] = v0
for t in range(1, M + 1):
    ran = np.dot(cho_mat, ran_num[:, t, :])
    vh[t] = (vh[t - 1] + kappa * (theta - np.maximum(vh[t - 1], 0)) * dt
          + sigma * np.sqrt(np.maximum(vh[t - 1], 0)) * np.sqrt(dt)  
          * ran[1])
v = np.maximum(vh, 0)
```

```python
S = np.zeros_like(ran_num[0])
S[0] = S0
for t in range(1, M + 1):
    ran = np.dot(cho_mat, ran_num[:, t, :])
    S[t] = S[t - 1] * np.exp((r - 0.5 * v[t]) * dt +
                    np.sqrt(v[t]) * ran[0] * np.sqrt(dt))
```

#### Jump-Diffusion

```python
S0 = 100.
r = 0.05
sigma = 0.2
lamb = 0.75
mu = -0.6
delta = 0.25
T = 1.0
```

```python
M = 50
I = 10000
dt = T / M
rj = lamb * (np.exp(mu + delta**2/2) - 1)
S = np.zeros((M + 1, I))
S[0] = S0
sn1 = npr.randn(M+1, I)
sn2 = npr.randn(M+1, I)
poi = npr.poisson(lamb * dt, (M + 1, I))
for t in range(1, M + 1):
    expo = (r - rj - sigma**2/2) * dt + sigma * np.sqrt(dt) * sn1[t]
    mult = np.exp(mu + delta * sn2[t]) - 1
    S[t] = S[t-1] * (np.exp(expo) + mult * poi[t])
    S[t] = np.maximum(S[t], 0)
```

```python
def gen_sn(M, I, anti_paths = True, mo_match = True):
    ''' Function to generate random numbers for simulation.
    
    Parameters
    ==========
    M : int
        number of time intervals for discretization
    I : int
        number of paths to be simulated
    anti_paths: boolean
        use of antithetic variates
    mo_math : boolean
        use of moment matching
    '''
    if anti_paths == True:
        sn = npr.standard_normal((M + 1, int(I / 2)))
        sn = np.concatenate((sn, -sn), axis = 1)
    else:
        sn = npr.standard_normal((M + 1, I))
    if mo_match == True:
        sn = (sn - sn.mean()) / sn.std()
    return sn
```

## Valuation
contingent claim의 가치는 risk neutral/martingale measure 하에서 expected payoff를 할인한 것과 같다. risk neutral measure는 주식, 지수 등의 모든 위험 요인이 risk free rate에 따라 증가하도록 하는 확률 측도이다.
Fundamental Theorem of Asset Pricing에 따르면, absence of arbirage(무재정)와 risk neutral measure는 서로 동치이다.

####  European option

```python
S0 = 100
r = 0.05
sigma = 0.25
T = 1
I = 50000
def gbm_mcs_stat(K):
    sn = gen_sn(1, I)
    S_T = S0 * np.exp((r - sigma**2 / 2) * T 
                      + sigma * np.sqrt(T) * sn[1])
    h_T = np.maximum(S_T - K, 0)
    C0 = np.exp(-r * T) * np.mean(h_T) 
    return C0
```

```python
gbm_mcs_stat(105)
```

```python
M = 50
def gbm_mcs_dyna(K, option = 'call'):
    dt = T / M
    S = np.zeros((M + 1, I))
    S[0] = S0
    sn = gen_sn(M,I)
    for t in range(1, M +1):
        S[t] = S[t-1] * np.exp((r - sigma**2 / 2) * dt 
                          + sigma * np.sqrt(dt) * sn[t])
    
    if option == 'call':
        h_T = np.maximum(S[-1] - K, 0)
    else: # Put Option
        h_T = np.maximum(K - S[-1], 0)
    
    C0 = np.exp(-r*T) * np.mean(h_T)
    return C0
```

### American Option
옵션 가치 계산과 관련하여 최적 정지 문제(Optimal stopping problem)을 풀어야 한다. 시뮬레이션 상에서는 이산 시간 그리드를 사용하여 풀 수 있다. 이렇게 하면 버뮤다 옵션 문제를 푸는 셈이지만, 시간 그리드의 간격을 줄이면 아메리칸 옵션으로 수렴한다.

```python
T
```

```python
def gbm_mcs_amer(K, option = 'call'):
    '''LSM(Least-Squares Monte Carlo) algorithm과 
    BSM Monte Carlo를 통한 American Option Valuation'''
    dt = T / M
    df = np.exp(-r * dt)
    
    # Stock Simulation
    S = np.zeros((M + 1, I))
    S[0] = S0
    sn = gen_sn(M, I)
    for t in range(1, M + 1):
        S[t] = S[t-1] * np.exp((r - sigma**2/2) * dt
                              + sigma * np.sqrt(dt) * sn[t])
    
    ## payoff
    if option == 'call':
        h = np.maximum(S - K, 0)
    else:
        h = np.maximum(K - S, 0)
    
    ## LSM algorithm
    V = np.copy(h)
    for t in range(M - 1, 0, -1):
        reg = np.polyfit(S[t], V[t + 1] * df, 7)
        C = np.polyval(reg, S[t])
        V[t] = np.where(C > h[t], V[t + 1] * df, h[t])
    
    ## MC estimation
    C0 = df * 1/I * np.sum(V[1])
    return C0
```

## Risk Measures

### Value-at-Risk
가장 널리 사용되며 가장 논란이 많은 위험 척도이다. 직관적이기 때문에 실무에서는 사랑받지만, tail risk를 제대로 파악하지 못한다는 점에서 많은 비판을 받음. VaR은 어떤 확률적인 신뢰도 하에서 특정 시간 내에 발생할 수 있는 화폐($, euro, GBP, ...)로 환산한 (포트폴리오나 단일 포지션의) 손실 가능 금액을 말한다.

eg. 1M 달러 주식 포지션의 99퍼센트 신뢰도 30일 VaR = 50K 달러. 30일 동안 포지션에서 30일 후 발생 가능한 손실은 99퍼센트 확률로 50K 달러를  넘지 않는다. 즉, 50K 달러 이상의 손실을 입을 확률이 1퍼센트라는 의미일 뿐이다.

```python
S0 = 100
r = 0.05
sigma = 0.25
T = 30 / 365
I = 10000
ST = S0 * np.exp((r - sigma**2 / 2) * T
                + sigma * np.sqrt(T) * npr.randn(I))
```

```python
R_gbm = np.sort(ST - S0)
```

```python
plt.hist(R_gbm, bins = 50)
plt.xlabel('absolute return')
plt.ylabel('frequency')
plt.title('Absolute returns of geometric Brownian motion (30d)')
plt.grid(True)
```

### Credit Value Adjustments
신용 VaR에서 파생. credit risk를 고려한 risk measure. 파산의 경우에는 PD와 loss level의 가정이 필요.

```python
S0 = 100
r = 0.05
sigma = 0.2
T = 1
I = 100000
ST = S0 * np.exp((r - sigma**2 / 2) * T
                + sigma * np.sqrt(T) * npr.randn(I))
```

```python
L = 0.5 # loss level
p = 0.01 # probability of default(PD)
```

```python
D = npr.poisson(p * T, I)
D = np.where(D > 1, 1, D) # 부도는 한 번만 일어난다.
```

```python
## Future index discounted by risk free rate
PV = np.exp(-r*T) * (1 / I) * np.sum(ST)
PV
```

```python
# Credit VaR
CVaR = np.exp(-r*T) * (1 / I) * np.sum(ST * L * D)
CVaR
```

```python
# 신용 위험을 조정한 자산의 현재 가치
S0_CVA = np.exp(-r*T) * (1 / I) * np.sum(ST * (1 - L * D))
S0_CVA
```

```python
# 현재 자산 가치에서 신용 VaR을 뺀 값
S0_adj = S0 - CVaR
S0_adj
```

```python
# 부도 횟수
np.count_nonzero(L * D * ST)
```

```python
plt.hist(L * D * ST, bins = 50)
plt.xlabel('loss')
plt.ylabel('frequency')
plt.grid(True)
plt.ylim(ymax = 150)
plt.title('Losses due to risk-neutrally expected default (stock)')
```

```python
K = 100
hT = np.maximum(ST - K, 0)
C0 = np.exp(-r*T) * (1 / I) * np.sum(hT)
C0
```

```python
CVaR = np.exp(-r*T) * (1 / I) * np.sum(L * D * hT)
CVaR
```

```python
# 조정된 옵션 가치
C0_CVA = np.exp(-r*T) * (1 / I) * np.sum((1 - L * D) * hT)
C0_CVA
```

```python
# 옵션의 경우 손실 횟수
np.count_nonzero(L * D * hT)
```
